```python
def IQR(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    res = Q3 - Q1
    return res

funcs =  ['median', IQR]   

evaluation_data['NumberOfStatesExceeded'] = evaluation_data['NumberOfStatesExceeded'].astype(int)
evaluation_data['QueryLengthExceeded'] = evaluation_data['QueryLengthExceeded'].astype(int)
```

```python
columns_to_compute = ['Time (s)', 'LastTokenQuery', 'EquivalenceQuery',  
                      'Number of Extracted States', 'NumberOfStatesExceeded', 'QueryLengthExceeded',
                      'Tree Depth', 'Inner Nodes', 'Accuracy',
                      'Unknown Results Percentage']



agg_dict =  {col:funcs for col in columns_to_compute}
grouped_multiple = evaluation_data.groupby(['Algorithm', 'Instance', 'MaxQueryLengthBound']).agg(agg_dict)
```

```python
col_names = []
for col in columns_to_compute:
    col_names.append(col+'_median')
    col_names.append(col+'_IQR')
grouped_multiple.columns = col_names
grouped_multiple.reset_index(inplace = True)
```
